```python
import pandas as pd
import numpy as np
```

```python
clean_jokes = pd.read_csv('only_true_jokes_CSVs/selection/clean_jokes.csv')
```

```python
rJokes_complete = pd.read_csv('only_true_jokes_CSVs/selection/rJokes_complete.csv')
```

```python
clean_jokes
```

```python
rJokes_complete
```

```python
rJokes_complete = rJokes_complete.rename(columns={"joke": "sentence"})
```

```python
rJokes_complete
```

```python
clean_jokes["sentence"] = clean_jokes["sentence"].str.strip().str.lower()
rJokes_complete["sentence"] = rJokes_complete["sentence"].str.strip().str.lower()
```

```python
clean_jokes_with_score = pd.merge(clean_jokes, rJokes_complete[['sentence', 'score']], on="sentence", how="left")
```

```python
clean_jokes_with_score
```

```python
clean_jokes_with_score_dropna = clean_jokes_with_score.dropna(subset=['score'])
```

```python
clean_jokes_with_score_dropna
```

```python
clean_jokes_with_score_dropna_remove_dups = clean_jokes_with_score_dropna.drop_duplicates()
```

```python
clean_jokes_with_score_dropna_remove_dups
```

```python
final_clean_df = clean_jokes_with_score_dropna_remove_dups.reset_index(drop=True)
```

```python
final_clean_df
```

## Save the new dataset

```python
final_clean_df.to_csv("only_true_jokes_CSVs/selection/final_clean_jokes_with_rating_and_test_sentences.csv", encoding='utf-8', index=False)
```
