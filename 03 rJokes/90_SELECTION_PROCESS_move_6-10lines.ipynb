```python
import os
import shutil

# Folder containing the text files
folder_path = "only_true_jokes_CSVs/selection"
short_folder = folder_path+"/6-10"

```

```python
print(short_folder)
```

```python
print (os.listdir(folder_path))
```

```python
# Create the "short" subdirectory if it doesn't exist
os.makedirs(short_folder, exist_ok=True)
```

```python

# Get all text files with not more than 6 lines and move them
for file_name in os.listdir(folder_path):
    file_path = folder_path+"/"+file_name

    # Check if it's a text file
    if file_name.endswith(".csv") and os.path.isfile(file_path):
        try:
            # Read the file and count lines
            with open(file_path, 'r') as file:
                lines = file.readlines()
                if len(lines) > 6 and len(lines) < 10:
                    # Move the file to the "short" folder
                    shutil.move(file_path, short_folder+"/"+file_name)
                    print(f"Moved: {file_name}")
        except Exception as e:
            print(f"Could not process {file_name}: {e}")

print("Finished moving short files.")
```

## Remove the last line from each file

```python
import os

# Path to the folder containing text files
folder_path = short_folder

# Iterate through all files in the directory
for file_name in os.listdir(folder_path):
    file_path = folder_path+"/"+file_name

    # Process only text files
    if file_name.endswith(".csv") and os.path.isfile(file_path):
        try:
            # Open the file and read all lines
            with open(file_path, 'r') as file:
                lines = file.readlines()

            # Write back the lines except the last one
            with open(file_path, 'w') as file:
                file.writelines(lines[:-1])

            print(f"Processed: {file_name}")

        except Exception as e:
            print(f"Could not process {file_name}: {e}")

print("Finished processing all files.")
```
