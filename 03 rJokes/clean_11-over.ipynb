```python
import os
folder_path = "only_true_jokes_CSVs/selection/11-over-cleaned"

```

```python


# Iterate through all files in the directory
for file_name in os.listdir(folder_path):
    file_path = folder_path+"/"+file_name

    # Process only text files
    if file_name.endswith(".csv") and os.path.isfile(file_path):
        try:
            # Open the file and read all lines
            with open(file_path, 'r') as file:
                lines = file.readlines()

            # Find the line containing "CSV" and remove all lines up to and including it
            new_lines = []
            found_csv = False
            for i, line in enumerate(lines):
                if "CSV" in line:
                    found_csv = True
                    new_lines = lines[i + 1:]  # Keep lines after the "CSV" line
                    break

            if found_csv:
                # Write the remaining lines back to the file
                with open(file_path, 'w') as file:
                    file.writelines(new_lines)
                print(f"Processed: {file_name}")
            else:
                print(f"No 'CSV' found in {file_name}, file unchanged.")

        except Exception as e:
            print(f"Could not process {file_name}: {e}")

print("Finished processing all files.")
```
