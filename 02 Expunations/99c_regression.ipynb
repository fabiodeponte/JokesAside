```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from math import sqrt
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('distances_single_tokens_colbert_12feb.csv', encoding='utf-8')
```

```python
df
```

```python
from sklearn.model_selection import train_test_split
```

### Remove NaNs

```python
rows_with_nans = df[df.isna().any(axis=1)]
```

```python
rows_with_nans
```

```python
data = df[['euclidean_distance', 'cosine_similarity']]
target = df['funniness_average']
#target = df['filtered_funniness_average']
```

```python
data_train, data_test, target_train, target_test = train_test_split(data, target, test_size=0.2, shuffle=True)
```

```python
data_train
```

```python
target_train
```

```python
model = LinearRegression()
model.fit(data_train, target_train)
```

```python
print("Coefficients:", model.coef_)
print("Intercept:", model.intercept_)
```

```python
predictions = model.predict(data_test)
score = model.score(data_test, target_test)
print("R^2 score:", score)
```

```python
mse = mean_squared_error(target_test, predictions)
```
