```python
import numpy as np
import pandas as pd
import ast
from sentence_transformers import util
import matplotlib.pyplot as plt
```

```python
human = pd.read_csv('human_artificial_minilm_openai_embeddings.csv')
```

```python
human
```

```python
for i in range(1):
    x = human.x[i]
    y = human.y[i]
    z = human.z[i]
    xz = human.x[i] + " " + human.z[i]
    yz = human.y[i] + " " + human.z[i]
    xyz = "I like my " + human.x[i] + " like I like my " + human.y[i] + ", " + human.z[i]

    print ("x:", x)
    print ("y:", y)
    print ("z:", z)
    print ("xz:", xz)
    print ("yz:", yz)
    print ("xyz:", xyz)
```

```python
'''
COLUMN NUMBERS
x - 0
y - 1
z - 2
score - 3
incomprehensibleMarkings - 4
offensiveMarkings - 5
average_score  - 6
emb_minilm_x   - 7
emb_minilm_y   - 8	
emb_minilm_z   - 9
emb_minilm_xz  - 10
emb_minilm_yz  - 11	
emb_minilm_xyz - 12
'''

```

# MINILM

### Difference between similarities xz-z and yz-z

```python
human['diff_minilm_xz-z_yz-z'] = None
human['diff_minilm_xz-z_yz-z'] = human.apply(
    lambda row: abs(
        util.cos_sim(ast.literal_eval(row['emb_minilm_xz']), ast.literal_eval(row['emb_minilm_z'])).item() - 
        util.cos_sim(ast.literal_eval(row['emb_minilm_yz']), ast.literal_eval(row['emb_minilm_z'])).item()
    ),
    axis=1
)
```

```python
print ("Example:")
print ("xz:", xz)
print ("yz:", yz)
print ("z:", z)

plt.scatter(human['diff_minilm_xz-z_yz-z'], human['average_score'], label="Difference of cosine similarity", marker="s", color="b")

# Add labels, title, and legend
plt.ylabel("Score")
plt.xlabel("Difference of cosine similarity")
plt.title("Difference of cosine similarity against score - XZ-Z and YZ-Z")
plt.legend()
plt.grid()

# Show plot
plt.show()
```

### SYMMETRY - Difference between similarities x-z and y-z

```python
human['minilm_symmetry'] = None
human['minilm_symmetry'] = human.apply(
    lambda row: abs(
        util.cos_sim(ast.literal_eval(row['emb_minilm_x']), ast.literal_eval(row['emb_minilm_z'])).item() - 
        util.cos_sim(ast.literal_eval(row['emb_minilm_y']), ast.literal_eval(row['emb_minilm_z'])).item()
    ),
    axis=1
)
```

```python
print ("Example:")
print ("x:", x)
print ("y:", y)
print ("z:", z)

plt.scatter(human['minilm_symmetry'], human['average_score'], label="minilm Symmetry", marker="s", color="b")

# Add labels, title, and legend
plt.ylabel("Score")
plt.xlabel("Difference of cosine similarity")
plt.title("Difference of cosine similarity against score - X-Z and Y-Z")
plt.legend()
plt.grid()

# Show plot
plt.show()
```
