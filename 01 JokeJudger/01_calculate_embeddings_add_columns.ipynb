# Calculate embeddings

### MODEL all-MiniLM-L6-v2

```python
from sentence_transformers import SentenceTransformer, util
import openai
```

```python
def sentence2embedding_minilm(sentence):

    # Load the all-MiniLM-L6-v2 model
    model = SentenceTransformer('sentence-transformers/all-MiniLM-L6-v2')
    return model.encode(sentence, convert_to_tensor=True)

    
```

### MODEL OpenAi

```python
# Set your OpenAI API key
# openai.api_key = ...

def sentence2embedding_openai(text, model="text-embedding-3-small"):
    """
    Fetch the embedding for a given text using OpenAI's updated embedding API.
    """
    response = openai.embeddings.create(
        input=text,
        model=model
    )
    # Access the embeddings from the response's 'data' attribute
    return response.data[0].embedding
    #return np.array(response['data'][0]['embedding'])
```

### Test: the the models return different results

```python
sentence1 = "The quick brown fox jumps over the lazy dog."
sentence2 = "jumps"

embedding_sentence1 = sentence2embedding_minilm(sentence1)
embedding_sentence2 = sentence2embedding_minilm(sentence2)
similarity = util.cos_sim(embedding_sentence1, embedding_sentence2)
print(f" Model MiniLM all-MiniLM-L6-v2 - Cosine Similarity between '{sentence1}' and '{sentence2}': {similarity.item():.4f}")

embedding_sentence1 = sentence2embedding_openai(sentence1)
embedding_sentence2 = sentence2embedding_openai(sentence2)
similarity = util.cos_sim(embedding_sentence1, embedding_sentence2)
print(f" Model OpenAi text-embedding-3-small - Cosine Similarity between '{sentence1}' and '{sentence2}': {similarity.item():.4f}")

```

### Add a column with cosine similarities MiniLM

```python
human['emb_minilm_x'] = None  # Initialize column with None
human['emb_minilm_x'] = human['emb_minilm_x'].astype(object)  # Ensure it's object type

human['emb_minilm_y'] = None  # Initialize column with None
human['emb_minilm_y'] = human['emb_minilm_y'].astype(object)  # Ensure it's object type

human['emb_minilm_z'] = None  # Initialize column with None
human['emb_minilm_z'] = human['emb_minilm_z'].astype(object)  # Ensure it's object type

human['emb_minilm_xz'] = None  # Initialize column with None
human['emb_minilm_xz'] = human['emb_minilm_xz'].astype(object)  # Ensure it's object type

human['emb_minilm_yz'] = None  # Initialize column with None
human['emb_minilm_yz'] = human['emb_minilm_yz'].astype(object)  # Ensure it's object type

human['emb_minilm_xyz'] = None  # Initialize column with None
human['emb_minilm_xyz'] = human['emb_minilm_xyz'].astype(object)  # Ensure it's object type


```

```python
for i in range(len(human)):
    x = human.x[i]
    y = human.y[i]
    z = human.z[i]
    xz = human.x[i] + " " + human.z[i]
    yz = human.y[i] + " " + human.z[i]
    xyz = "I like my " + human.x[i] + " like I like my " + human.y[i] + ", " + human.z[i]

    human.at[i, 'emb_minilm_x'] = sentence2embedding_minilm(x).tolist()
    human.at[i, 'emb_minilm_y'] = sentence2embedding_minilm(y).tolist()
    human.at[i, 'emb_minilm_z'] = sentence2embedding_minilm(z).tolist()
    human.at[i, 'emb_minilm_xz'] = sentence2embedding_minilm(xz).tolist()
    human.at[i, 'emb_minilm_yz'] = sentence2embedding_minilm(yz).tolist()
    human.at[i, 'emb_minilm_xyz'] = sentence2embedding_minilm(xyz).tolist()

    print(i, "-", xyz, f"- {human.average_score[i]:.2f}")
```
